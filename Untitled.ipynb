# Linear Regression With One Variable using Scikitlearn

```python
X = pd.DataFrame(np.c_[car_data_df['Year']], columns = ['Year'])
Y = car_data_df['Selling_Price']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#entrainement du modÃ¨le
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
 
lmodellineaire = LinearRegression()
lmodellineaire.fit(X_train, Y_train)
Y_pred = lmodellineaire.predict(X_test)
plt.scatter(Y_test, Y_pred)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, Y_pred)
print(mse,lmodellineaire.intercept_,lmodellineaire.coef_)
```

```python
x = car_data_df['Year']
y = car_data_df['Selling_Price']
```

```python
plt.plot(x, y, 'o', label='original data')
plt.plot(x, lmodellineaire.intercept_ + lmodellineaire.coef_*x, 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
# Evaluation du training set
from sklearn.metrics import r2_score
y_train_predict = lmodellineaire.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)
 
print('La performance du modÃ¨le sur la base dapprentissage')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
print('\n')
 
# model evaluation for testing set
y_test_predict = lmodellineaire.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
 
print('La performance du modÃ¨le sur la base de test')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
```

# Linear Regression With One Variable Using statsmodels 

```python
# For statistics. Requires statsmodels 5.0 or more
from statsmodels.formula.api import ols
# Analysis of Variance (ANOVA) on linear models
from statsmodels.stats.anova import anova_lm
data = pd.DataFrame({'Year': car_data_df['Year'], 'Selling_Price': car_data_df['Selling_Price']})
# Fit the model
model = ols("y ~ x", data).fit()
model
```

```python
# Print the summary
print(model.summary())
# Peform analysis of variance on fitted linear model
anova_results = anova_lm(model)
print('\nANOVA results')
print(anova_results)
```

```python
# Retrieve the parameter estimates
offset, coef = model._results.params
plt.plot(x, x*coef + offset, c = 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x,y)
plt.show()
```

# 2 Variables(Year and Present Price) Linear regression With Using SkLearn

```python
######################################## Data preparation #########################################
from sklearn import linear_model
X = car_data_df2[['Year', 'Present_Price']].values
Y = car_data_df2['Selling_Price']

######################## Prepare model data point for visualization ###############################
x = X[:, 0]
y = X[:, 1]
z = Y
x_pred = np.linspace(2012, 2019, 15)   # range of porosity values
y_pred = np.linspace(0, 20, 15)  # range of brittleness values
xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)
model_viz = np.array([xx_pred.flatten(), yy_pred.flatten()]).T

################################################ Train #############################################

ols = linear_model.LinearRegression()
model = ols.fit(X, Y)
predicted = model.predict(model_viz)

############################################## Evaluate ############################################

r2 = model.score(X, Y)

############################################## Plot ################################################

plt.style.use('default')

fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')

axes = [ax1, ax2, ax3]

for ax in axes:
    ax.plot(x, y, z, color='k', zorder=15, linestyle='none', marker='o', alpha=0.5)
    ax.scatter(xx_pred.flatten(), yy_pred.flatten(), predicted, facecolor=(0,0,0,0), s=20, edgecolor='#70b3f0')
    ax.set_xlabel('Year', fontsize=12)
    ax.set_ylabel('Present_Price', fontsize=12)
    ax.set_zlabel('Selling_Price', fontsize=12)
    ax.locator_params(nbins=4, axis='x')
    ax.locator_params(nbins=5, axis='x')
    
    
ax1.view_init(elev=8, azim=160)
ax2.view_init(elev=45, azim=-30)
ax3.view_init(elev=17, azim=25)

fig.suptitle('$R^2 = %.2f$' % r2, fontsize=20)

fig.tight_layout()
```

# Prediction of Selling Price

```python
# function for prediction
def prediction(year, presentprice):
    predictedPrice = ols.predict([[year, presentprice]])
    return predictedPrice[0].round(2)


# taking parameters from user
year=int(input('Quel modÃ¨le devrait Ãªtre votre voiture?(Year, ex: 2014) : '))
presentprice=float(input('Que devrait valoir votre voiture maintenant?(Present_Price, ex: 14.5) : '))
prediction(year, presentprice)
```
